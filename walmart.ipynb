## Business problems

```python
# find different payment method and number of transactions ,n of qty sold
```

```python
query="select payment_method ,count(*) as n_payments, SUM(quantity) as n_qty_sold from walmart GROUP BY payment_method"
sql_query(query)
```

```python
# identify the highest-rated category in each branch , displaying the branch,category avg rating
```

```python
query="select category ,MAX(rating) as highest ,AVG(rating) as avg_rating from walmart GROUP BY category"
sql_query(query)
```

```python
query="SELECT Branch, category, avg_rating FROM ( SELECT Branch,category,AVG(rating) AS avg_rating,RANK() OVER(PARTITION BY Branch ORDER BY AVG(rating) DESC) AS rankK FROM walmart GROUP BY 1,2) AS ranked WHERE rankK = 1"
sql_query(query)
```

```python
# identify the busiest day for each branch based on the number of transactions
```

```python
query="select Branch , day_name,n_transactions from (select Branch , dayname(date) as day_name , count(*) as n_transactions,rank() over(partition by Branch order by count(*) desc) as r from walmart group by Branch , day_name) as ran where r=1"
sql_query(query)
```

```python
# Calculate the total quantity of items sold per payment method
```

```python
query="select payment_method ,SUM(quantity) from walmart GROUP BY payment_method"
sql_query(query)
```

```python
# Determine the average, minimum, and maximum rating of categories for each city
```

```python
query="select category , city ,MAX(rating),MIN(rating),AVG(rating) from walmart GROUP BY category , city"
sql_query(query)
```

```python
# Calculate the total profit for each category
```

```python
query="select category, SUM(unit_price * quantity * profit_margin) from walmart GROUP BY category"

sql_query(query)
```

```python
# Determine the most common payment method for each branch
```

```python
query="WITH cte AS(select Branch ,payment_method,count(*), rank() over(partition by Branch order by count(*) DESC) as r from walmart GROUP BY Branch,payment_method) SELECT Branch, payment_method AS preferred_payment_method FROM cte WHERE r = 1"

sql_query(query)
```

```python
# Categorize sales into Morning, Afternoon, and Evening shifts
```

```python
query="select Branch, CASE WHEN HOUR(time) < 12 THEN 'Morning' WHEN HOUR(time) BETWEEN 12 AND 17 THEN 'Afternoon' ELSE 'Evening' END AS shift ,COUNT(*) AS num_invoices FROM walmart GROUP BY Branch, shift ORDER BY Branch, num_invoices DESC"
sql_query(query)
```
